```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('fact_user_session.csv')
df.head()
```

```python
df['SESSION_START'] =  pd.to_datetime(df['SESSION_START'], format='%Y-%m-%d %H:%M:%S.%f')
```

```python
df['SESSION_END'] =  pd.to_datetime(df['SESSION_END'], format='%Y-%m-%d %H:%M:%S.%f')
```

```python
df['SESSION_START'][0].strftime("%A")
```

```python
df['SESSION_START'][0].time()
```

```python
df['Day'] = df.apply(lambda row: row.SESSION_START.strftime("%A"), axis=1)
df
```

```python
df['Time'] = df.apply(lambda row: row.SESSION_START.time(), axis=1)
df
```

```python
df.info()
```

```python
df['SESSION_START'][0].time()
```

```python
def StartLaterThan2(x):
    dt_object = x.time()
    if dt_object > datetime.time(14, 0, 0, 0):
        y = True
    else:
        y = False
    return y

df['Starts_Later_Than_2'] = df['SESSION_START'].apply(StartLaterThan2)
df
```

```python
def EndsAfter2(x):
    dt_object = x.time()
    if dt_object > datetime.time(14, 0, 0, 0):
        y = True
    else:
        y = False
    return y

df['Ends_After_2'] = df['SESSION_END'].apply(EndsAfter2)
df
```

```python
df['isActive'] = df.apply(lambda row: row.Starts_Later_Than_2 + row.Ends_After_2, axis=1)
df
```

```python
df_chosen = df.loc[df['isActive'] != 0]
df_chosen
```

```python
df_chosen['Day'].value_counts()
```

```python
df['Day'].value_counts()
```

```python
active_users = pd.DataFrame(df_chosen['Day'].value_counts())
active_users = active_users.reindex(['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday'])
```

```python
active_users.reset_index(inplace=True)
active_users
```

```python
active_users.rename(columns = {'index':'Day', 'Day': 'Active Users Between 14.00 and 23.59'}, inplace = True)
active_users
```

```python
line = active_users.plot.line(x='Day', y='Active Users Between 14.00 and 23.59')
```
